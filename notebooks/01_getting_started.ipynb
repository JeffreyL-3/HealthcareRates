## 1. Import libraries

An `import` makes a library available in the notebook. The short names `pd`, `sns`, and `plt` are standard Python conventions.

```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

## 2. Load a manageable sample

The full CSV is large, so begin with 10,000 rows. Remove `nrows=10_000` later when you want the complete dataset.

```python
data_path = Path("../data/Q5107 All National General.csv")

rates = pd.read_csv(data_path, nrows=10_000, low_memory=False)
rates.head()
```

## 3. Understand the data

These are good first questions for any new data frame.

```python
print(f"Rows: {len(rates):,}")
print(f"Columns: {len(rates.columns)}")
rates.info()
```

```python
# Count missing values, showing the columns with the most missing data first.
rates.isna().sum().sort_values(ascending=False).head(10)
```

## 4. Summarize rates by state

This is similar to `group_by(STATE) |> summarize(...)` in dplyr.

```python
state_summary = (
    rates.dropna(subset=["STATE", "RATE"])
    .groupby("STATE", as_index=False)
    .agg(
        observations=("RATE", "size"),
        median_rate=("RATE", "median"),
    )
    .sort_values("median_rate", ascending=False)
)

state_summary.head(10)
```

## 5. Make a chart

```python
plot_data = state_summary.head(15).sort_values("median_rate")

sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 7))
sns.barplot(data=plot_data, x="median_rate", y="STATE", color="#2878B5")
plt.title("Median negotiated rate by state")
plt.xlabel("Median rate ($)")
plt.ylabel("State")
plt.show()
```
